```python
#Merging DataFrames is based on the same values on both dataframe
#first creating 3 data frames 

# DataFrame 1: Clothing Items
data1 = {
    'ItemID': [1, 2, 3],
    'ItemName': ['Abaya', 'Hijab', 'Jilbab'],
    'DesignerID': [101, 102, 101]
}

df_cloth1 = pd.DataFrame(data1)

# DataFrame 2: Designers
data2 = {
    'DesignerID': [101, 102],
    'DesignerName': ['Designer A', 'Designer B']
}

df_cloth2 = pd.DataFrame(data2)

# DataFrame 3: Clothing Items
data3 = {
    'ItemID': [4, 5,9],
    'ItemName': ['Kaftan', 'Tunic','Jilbab'],
    'DesignerID': [103, 104,105]
}

df_cloth3 = pd.DataFrame(data3)


print("DataFrame clothes 1:\n", df_cloth1)
print("\nDataFrame clothes 2:\n", df_cloth2)
print("\nDataFrame clothes 3:\n", df_cloth3)
```

```python
Merged_dataFrame=df_cloth1.merge(df_cloth2,on='DesignerID')
print(Merged_dataFrame)

#There should be same values in the columns on the base of which you are merging them
Merged_dataFrame2=Merged_dataFrame.merge(df_cloth3,on='DesignerID')
print(Merged_dataFrame2)
```

```python
#Concatenation vertically
Concatenated_DataFrame=pd.concat([Merged_dataFrame,df_cloth3], ignore_index=True,axis=0)
print(Concatenated_DataFrame)
Concatenated_DataFrame=pd.concat([Merged_dataFrame,df_cloth3], ignore_index=True,axis=1)
print(Concatenated_DataFrame)
```

```python
#Concatenation with keys to create a Multiindex
#You can merged more then one dataframe
Concate_multiindex= pd.concat([Merged_dataFrame,df_cloth3], keys=['Modest_Clothing','Medieval_Clothing'],axis=0)
Concate_multiindex
```
