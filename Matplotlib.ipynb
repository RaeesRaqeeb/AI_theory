```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4],[5,6,7,4])
plt.ylabel('some numbers')
plt.legend("c")
plt.show()
```

```python
x = [1, 2, 3, 4] # x-axis values
y = [2, 4, 6, 8] # y-axis values
plt.plot(x, y, marker='o', linestyle=':', color='b') # Marker, linestyle, and
# color customization
plt.xlabel('X Axis') # X-axis label
plt.ylabel('Y Axis') # Y-axis label
plt.title('Customized Line Chart') # Plot title
plt.grid(True) # Show gridlines
plt.show() # Display the plot
```

```python
plt.plot(x, y, marker='o',markeredgecolor='#2b8a28',markerfacecolor='none',linestyle='--', color='#4af08a',mew=1,markersize=15) # Marker, linestyle,
# and color customization
plt.xlabel('X Axis') # X-axis label
plt.ylabel('Y Axis') # Y-axis label
plt.title('Customized Line Chart') # Plot title
plt.grid(True) # Show gridlines
plt.show() # Display the plot
```

```python
# Data for bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [4, 7, 1, 6, 10]

# Create a bar chart

plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

```python
data = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.subplot(1,3,1)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Data for bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [4, 7, 1, 6, 10]
plt.subplot(1,3,2)
# Create a bar chart

plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

plt.subplot(1,3,3)
plt.plot(x, y, marker='o',markeredgecolor='#2b8a28',markerfacecolor='none',linestyle='--', color='#4af08a',mew=1,markersize=15) # Marker, linestyle,
# and color customization
plt.xlabel('X Axis') # X-axis label
plt.ylabel('Y Axis') # Y-axis label
plt.title('Customized Line Chart') # Plot title
plt.grid(True) # Show gridlines
plt.show() # Display the plot
```
